# BeautifulSoup

```python
import requests
url = "https://www.frasesmania.com/"
r = requests.get(url, headers={'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko)'})
print(r.content)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(r.content, 'html5lib')
print(soup.prettify())
```

```python
table = soup.find('div', attrs={'id': 'content'})
print(table.prettify())
```

```python
for row in table.findAll( 'div', attrs={'class':'uagb-post__excerpt'}) : print(row.prettify())
```
